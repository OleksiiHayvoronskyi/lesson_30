**Завдання 2**. $\\ $ 
*Випадкова величини $ Х $ розподілена рівномірно*. 
* Знайти щільність її розподілу, якщо $ P(X \ge 3)\mathop 0,4 $, а математичне сподівання дорівнює $ M(X) = 2 $.

Оскільки випадкова величина $ X $ рівномірно розподілена, то її функція щільності визначається як: $ \\ $

$ f(x) = \frac {1} {b \mathop - a} $, для $ a \le x \le b $, $ \\ $

де $ a $ і $ b $ верхня та нижня межі розподілу.

$ P(x \ge 3) = \int^b _3 f(x)dx $: $ \\ $

$ \int^b _3 \frac {1} {b - a}dx = \frac {b - 3} {b - a} = 0,4 $

Знаючи математичне сподівання, обчислюємо середнє значення: $ \\ $

$ M(x) = \frac {a \mathop + b} {2} = 2 $.

**Знаходимо** $ a $ і $ b $: $ \\ $
$ a + b = 4 $.

$ \frac {b \mathop - 3} {b \mathop - a} = 0,4 $.

$ b = 3 - 0,4 = 2,6 $.

$ a = 4 - 2,6 = 1,4 $.

**РЕЗУЛЬТАТ**. $ f(x) = \frac {1} {2,6 \mathop - 1,4} = 0,83 $, для $ 1,4 \le x \le 2,6 $.
